```python
#read from json and transform the data.
import json
import pandas as pd
import os
pd.set_option('display.max_columns', None)
os.chdir("/Users/Lenovo/")
```

```python
f = open('politicalBots.json') 
data = json.load(f) 
dfs = []
```

```python
for i in range(len(data)):
    #if i!= 4589 and i!=10916 and i!=12308: #for celebrity
    df = pd.DataFrame.from_dict(data[i]['user'])
    dfs.append(df)
df = pd.concat(dfs,sort=True).reset_index()
df
```

```python
df = df.drop_duplicates(subset=['id']).sort_values(by='id')
df
```

```python
df2 = pd.read_csv("vendorPurchased.tsv", sep='\t')
df2 = df2.rename(columns={'2718436417':'test_id','human':'isGenuine'})
df2['knownBot'] = df2['isGenuine'].map({'human': int(0), 'bot': int(1)})
df2 = df2.drop(columns='isGenuine')
df2 
```

```python
df2 = df2.drop_duplicates(subset=['test_id']).sort_values(by='test_id')
df2
```

```python
lis1=[]
for i in df2['test_id']:
    lis1.append(i)
df_ = df[df.id.isin(lis1)].sort_values(by='id')
#print(test_score)
lis2 = []
for i in df['id']:
    lis2.append(i)
df2_ = df2[df2.test_id.isin(lis2)].sort_values(by='test_id')
df = df_
df2 = df2_
print(len(df),len(df2))
```

```python
j=0
l = []
for i in range(len(df['id'])):
    if list(df['id'])[i] != list(df2['test_id'])[i+j]:
        j += 1
        l.append(i)
df2.drop(l, inplace=True)
print(len(df),len(df2))
```
